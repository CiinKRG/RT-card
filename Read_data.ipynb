```python
import pandas as pd
from datetime import datetime, date 
url = "/content/data.csv"
df = pd.read_csv(url, error_bad_lines=False, delimiter=",")
```

```python
df.head()
```

```python
#Filtro
df_filter = df[(df.status_proceso == "NOK") & (df.no_celular != "")]
def age(born): 
    born = datetime.strptime(born, "%m/%d/%Y").date() 
    today = date.today() 
    return today.year - born.year - ((today.month,  
                                      today.day) <(born.month,  
                                                    born.day)) 
  
df_filter['Edad'] = df_filter['fecha_nacimiento'].apply(age) 
df_filter
```

```python
#filtro campos de interes bot
df_filter=df_filter.filter(items=['nombre', 'no_celular','Edad','counter','correo'])
df_filter
```

```python
#filtro edad
df_bot_50 = df_filter[(df_filter.Edad >= 55)]
#filtro edad menores de 55
df_bot = df_filter[(df_filter.Edad < 55)]
#Recibiran un correo antes de mandar mensaje por el bot
```
